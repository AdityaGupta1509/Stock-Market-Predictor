# Building a backtesting System

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(d, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, d.shape[0], step):
        train = d.iloc[0:i].copy()
        test = d.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(d, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
#%age of days stock went up or down
#formuls is value count divided by total no. of rows

predictions["Target"].value_counts() / predictions.shape[0]
```

# Adding additional predictors

```python
#creating a variety of rolling averages
#calculate the close mean price in last 2 days last trading week ie 5
#last 3 month ie 60 last year ie 250 and last 4 year ie 1000

horizons = [2,5,60,250,1000]
new_predictors = []    #hold new columns we create

for horizon in horizons:
    rolling_averages = d.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    d[ratio_column] = d["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    d[trend_column] = d.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors+= [ratio_column, trend_column]
```

```python
d = d.dropna(subset=d.columns[d.columns != "Tomorrow"])
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.6] = 1
    preds[preds <.6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(d, model, new_predictors)
```
